```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
group_labels = ['Gen Z', 'Millenial', 'Gen X', 'Boomer+']
groups = ['genZ', 'millenial', 'genX', 'boomer']

dfs_dict = {}
for g in groups:
    df = pd.read_csv('../prolific_results/'+g+'_pp_numeric.csv', index_col=False).iloc[2:].reset_index(drop=True)
    if (g == 'genZ'):
        df = df.iloc[5:].reset_index(drop=True)
    dfs_dict[g] = df
```

```python
# plot Q1

demo = 'genZ'
all_answer_frequencies = []
all_answer_frequencies_se = []
for demo in groups:
    # for one demographic
    df = dfs_dict[demo]
    df['Q1_ls'] = '[' + df['Q1'].astype(str)+']'
    df['Q1_ls'] = df['Q1_ls'].apply(eval)
    answer_frequencies = []
    answer_frequencies_se = []
    for answer in [1,2,3]:
        answer_count = np.sum(df['Q1_ls'].apply(lambda x: answer in x))
        answer_frequencies.append(answer_count)
        p = answer_count/len(df['Q1_ls'])
        se = 1.96*np.sqrt(p*(1-p))/np.sqrt(len(df['Q1_ls']))
        answer_frequencies_se.append(se)
    answer_frequencies = np.array(answer_frequencies)/50
    all_answer_frequencies.append(answer_frequencies)
    all_answer_frequencies_se.append(answer_frequencies_se)

# Example data
metrics = ['Search Engines', 'Generative\nSearch Engines', 'Chatbots\nwithout Citation']
demographics = group_labels
values = np.array(all_answer_frequencies)

# Bar width and positioning
bar_width = 0.2
x = np.arange(len(metrics))

# Create the plot
fig, ax = plt.subplots()

# Plot bars for each demographic
for i, demo in enumerate(demographics):
    ax.bar(x + i * bar_width, values[i], width=bar_width, label=demo, yerr=all_answer_frequencies_se[i])

# Customize the x-axis
ax.set_xticks(x + bar_width * 1.5)
ax.set_xticklabels(metrics)

# Add legend, labels, and title
ax.legend(title="Demographics")
ax.set_ylabel('Proportion of Respondents\nwho have used the platform category')
ax.set_xlabel('Platform category')
ax.set_title('Platform Use by Generation')

# Display the plot
plt.show()
print(all_answer_frequencies_se)
```

```python
# plot Q2
q_num = '6'
answer_choices = [1,2]
demographics = ['Gen Z','Millenial','Gen X','Boomer+']
metrics = ['high-stakes', 'low-stakes']
question = 'If you are seeking information that will influence your\ndecision-making in a ________ setting, which platform\nwould you rather use to find the information?'

all_answer_frequencies = []
all_answer_frequencies_se = []
for demo in groups:
    # for one demographic
    df = dfs_dict[demo]

    exposed_df = df[df['Q1_ls'].apply(lambda x: (1 in x) and (3 in x))]
    num_answered_se_6 = len(exposed_df[exposed_df['Q6']=='2'])
    num_answered_se_7 = len(exposed_df[exposed_df['Q7']=='2'])
    all_answer_frequencies.append([num_answered_se_6/len(exposed_df), num_answered_se_7/len(exposed_df)])

    p6 = num_answered_se_6/len(exposed_df)
    p7 = num_answered_se_7/len(exposed_df)
    se6 = 1.96*np.sqrt(p6*(1-p6))/np.sqrt(len(exposed_df))
    se7 = 1.96*np.sqrt(p7*(1-p7))/np.sqrt(len(exposed_df))

    all_answer_frequencies_se.append([se6, se7])

demographics = group_labels
values = np.array(all_answer_frequencies).T
all_answer_frequencies_se = np.array(all_answer_frequencies_se).T
# Bar width and positioning
x = np.arange(len(demographics))
bar_width = .34

# Create the plot
fig, ax = plt.subplots()

# Plot bars for each demographic
for i, met in enumerate(metrics):
    ax.bar(x + i * bar_width, values[i], width=bar_width, label=met, yerr=all_answer_frequencies_se[i])

# Customize the x-axis
ax.set_xticks(x+bar_width/2)
ax.set_xticklabels(demographics)

# Add legend, labels, and title
ax.set_ylabel('Proportion of Respondents\nwho prefer chatbots over search engines')
ax.set_xlabel('Demographic')
ax.set_title(question)
ax.legend(title="Setting")
plt.ylim(0,.65)
# Display the plot
plt.show()
all_answer_frequencies
```

```python
# plot Q6 aggregate across all demographics
answer_choices = [1, 2]
metrics = ['high-stakes', 'low-stakes']
question = 'If you are seeking information that will influence your\ndecision-making in a ________ setting, which platform\nwould you rather use to find the information?'

# Initialize variables for aggregated counts
total_exposed = 0
total_num_answered_se_6 = 0
total_num_answered_se_7 = 0
vec6_aggregate = []
vec7_aggregate = []

# Aggregate the data across all demographics
for demo in groups:
    df = dfs_dict[demo]

    exposed_df = df[df['Q1_ls'].apply(lambda x: (1 in x) and (3 in x))]
    num_answered_se_6 = len(exposed_df[exposed_df['Q6'] == '2'])
    num_answered_se_7 = len(exposed_df[exposed_df['Q7'] == '2'])
    
    total_exposed += len(exposed_df)
    total_num_answered_se_6 += num_answered_se_6
    total_num_answered_se_7 += num_answered_se_7

# Calculate the proportions
prop_se_6 = total_num_answered_se_6 / total_exposed
prop_se_7 = total_num_answered_se_7 / total_exposed

se6_aggregate = 1.96 * np.sqrt(prop_se_6*(1-prop_se_6)/total_exposed)
se7_aggregate = 1.96 * np.sqrt(prop_se_7*(1-prop_se_7)/total_exposed)

# Prepare data for the plot
values_aggregate = [prop_se_6, prop_se_7]
se_aggregate = [se6_aggregate, se7_aggregate]

# Create the plot
fig, ax = plt.subplots()

# Bar width and positioning
x = np.arange(len(metrics))  # Two bars for high-stakes and low-stakes

# Plot bars for aggregate data
ax.bar(x, values_aggregate, yerr=se_aggregate, label="Aggregated")

# Customize the x-axis
ax.set_xticks(x)
ax.set_xticklabels(metrics)

# Add legend, labels, and title
ax.set_ylabel('Proportion of Respondents\nwho prefer chatbots over search engines')
ax.set_xlabel('Setting')
ax.set_title(question)

# Display the plot
plt.ylim(0,.65)
plt.show()
```

```python
# plot Q2 by demographic
q_num = '5' # 2-5
answer_choices = [1,2,3]
metrics = ['Search Engine', 'Generative Search Engine', 'Chatbot']

questions = {'2':'If you have a question concerning how to handle a personal medical issue,\nwhich platform would you use to seek an answer?',
            '3':'If you have a question concerning topics in a friendly game of trivia,\nwhich platform would you use to seek an answer?',
             '4':'If you had a question concerning how to file your taxes,\nwhich platform would you use to seek an answer?',
             '5':'If you were trying to explain a complicated science topic to your third grade niece,\nwhich platform would you use to seek an answer?',
            }
question = questions[q_num]
all_answer_frequencies = []
all_answer_frequencies_se = []
for demo in groups:
    # for one demographic
    df = dfs_dict[demo]
    exposed_df = df
    exposed_df = df[df['Q1_ls'].apply(lambda x: (1 in x) and (3 in x))]
    num_answered_1 = len(exposed_df[exposed_df['Q'+q_num]=='1'])
    num_answered_2 = len(exposed_df[exposed_df['Q'+q_num]=='2'])
    num_answered_3 = len(exposed_df[exposed_df['Q'+q_num]=='3'])
    all_answer_frequencies.append([num_answered_1/len(exposed_df), num_answered_2/len(exposed_df), num_answered_3/len(exposed_df)])

    p1 = num_answered_1 / len(exposed_df)
    p2 = num_answered_2 / len(exposed_df)
    p3 = num_answered_3 / len(exposed_df)
    se1 = 1.96*np.sqrt(p1*(1-p1))/np.sqrt(len(exposed_df))
    se2 = 1.96*np.sqrt(p2*(1-p2))/np.sqrt(len(exposed_df))
    se3 = 1.96*np.sqrt(p3*(1-p3))/np.sqrt(len(exposed_df))

    all_answer_frequencies_se.append([se1, se2, se3])

# Example data

demographics = group_labels
values = np.array(all_answer_frequencies).T
all_answer_frequencies_se = np.array(all_answer_frequencies_se).T
# Bar width and positioning
x = np.arange(len(demographics))
bar_width = .3

# Create the plot
fig, ax = plt.subplots()

# Plot bars for each demographic
for i, met in enumerate(metrics):
    ax.bar(x + i * bar_width, values[i], width=bar_width, label=met, yerr=all_answer_frequencies_se[i])

# Customize the x-axis
ax.set_xticks(x+bar_width/2)
ax.set_xticklabels(demographics)

# Add legend, labels, and title
ax.set_ylabel('Proportion of Respondents\nwho prefer the specified platform')
ax.set_xlabel('Demographic')
ax.set_title(question)
ax.legend(title="Setting")
plt.ylim(0,1)
# Display the plot
plt.show()
```

```python
# Visualize reasons for preferring search engines over chatbots, or vice versa
# plot Q2
answer_choices = [1,2]
demographics = ['Gen Z','Millenial','Gen X','Boomer+']
metrics = ['...search engines...', 
           '...chatbots...']
question = 'What are your reasons to use ___________ when\nseeking information to answer a query?'
reasons_x_labels = ['Reliability of\nprovided answers',
                   'Convenience of\nobtaining an answer',
                   'Knowing where the\nanswer comes from',
                   'Entertainment of the \nunderlying technology',
                   'Variety of sources\nmade available',
                   'Seeking a highly\npersonalized answer']

all_answer_frequencies = []
all_answer_frequencies_se = []

num_answered_8 = np.zeros(6)
num_answered_9 = np.zeros(6)
vecs_8 = [[] for _ in range(6)]
vecs_9 = [[] for _ in range(6)]
total_num = 0
for demo in groups:
    # for one demographic
    df = dfs_dict[demo]

    exposed_df = df[df['Q1_ls'].apply(lambda x: (1 in x) and (3 in x))]
    curr_num_answered_8 = []
    curr_num_answered_9 = []
    for i, answer_key in enumerate(['1','2','3','5','6','7']):
        curr_num_answered_8.append(len(exposed_df[exposed_df['Q8'].str.contains(answer_key)]))
        curr_num_answered_9.append(len(exposed_df[exposed_df['Q9'].str.contains(answer_key)]))
        vecs_8[i].extend([1]*curr_num_answered_8[-1]+[0]*(len(exposed_df) - curr_num_answered_8[-1]))
        vecs_9[i].extend([1]*curr_num_answered_9[-1]+[0]*(len(exposed_df) - curr_num_answered_9[-1]))
        
    num_answered_8 += np.array(curr_num_answered_8) # each entry is the number of people who selected that box
    num_answered_9 += np.array(curr_num_answered_9)
    total_num += len(exposed_df) # the total number of people who responded

ses8 = [1.96*np.sqrt(n/total_num * (1-n/total_num))/np.sqrt(total_num) for n in num_answered_8]
ses9 = [1.96*np.sqrt(n/total_num * (1-n/total_num))/np.sqrt(total_num) for n in num_answered_9]

# sort by blue bar height
sorted_indices = np.argsort(-num_answered_8)
num_answered_8 = np.array(num_answered_8)[sorted_indices]
num_answered_9 = np.array(num_answered_9)[sorted_indices]
ses8 = np.array(ses8)[sorted_indices]
ses9 = np.array(ses9)[sorted_indices]
reasons_x_labels = np.array(reasons_x_labels)[sorted_indices]
all_answer_frequencies = [num_answered_8/total_num, num_answered_9/total_num]
all_answer_frequencies_se = [ses8, ses9]
# Example data

demographics = group_labels
values = np.array(all_answer_frequencies)

# Bar width and positioning
x = np.arange(6)
bar_width = .34

# Create the plot
fig, ax = plt.subplots()
fig.set_size_inches(5,2)
# Plot bars for each demographic
for i, met in enumerate(metrics):
    ax.bar(x + i * bar_width, values[i], width=bar_width, label=met, yerr=all_answer_frequencies_se)

# Customize the x-axis
ax.set_xticks(x+bar_width/2)
ax.set_xticklabels(['']*6)

for i, label in enumerate(reasons_x_labels):
    plt.text(
        x[i],
        -0.15,  # x position is from data, y position is manual offset
        label,
        ha='left',   # Adjust alignment
        rotation=315,
        rotation_mode='anchor'
    )

plt.xticks(rotation=315, ha='left', va='top')

# Add legend, labels, and title
ax.set_ylabel('Proportion of Respondents\nwho include the reason')
ax.set_title(question, pad=14)
ax.legend(bbox_to_anchor=(1.01,1.026), labelspacing=.25, fontsize=9)
plt.ylim(0,1)


# Display the plot
plt.savefig('figures/prolific_platform_preference_reasons', bbox_inches='tight', dpi=400)

plt.show()
```
